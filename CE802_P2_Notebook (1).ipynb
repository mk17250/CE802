# Applying PCA to data

```python
# set new x y for pca

# Set x and y 
x1 = data1.loc[:, data1.columns != 'Class'].to_numpy()
y1 = data1.loc[:, 'Class'].to_numpy()


# Scale data and apply PCA
y_pca = y1
scaler = StandardScaler()
scaler.fit(x1)
x_pca = scaler.transform(x1)
pca = PCA(n_components=2)
pca.fit(x_pca)
x_pca = pca.transform(x_pca)
```

```python
# Create new df containing principle components and target
principalDf = pd.DataFrame(data = x_pca
             , columns = ['principal component 1', 'principal component 2'])
finalDf = pd.concat([principalDf, pd.DataFrame(y_pca, columns = ['Class'])], axis=1)
finalDf
```

```python
# plot scatter of principle components 
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [0, 1]
colors = ['r', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Class'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

## SVM

```python
#set labels for new df 
data1['Class'] = data1['Class'].astype(int)
# data1
```

```python
# set new x,y for new df with nans still present 
x_train = data2.loc[:, data1.columns != 'Class'].to_numpy()
y_train = data2.loc[:, 'Class'].to_numpy()
print(x_train.shape, y_train.shape)
```

```python
imputer = SimpleImputer() #set imputer

# Choose cross-validation techniques for the inner and outer loops for nested cv
inner_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)
outer_cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)
```

```python
## PLEASE NOTE! PCA and scaling was explored and no PCA and StandardScaler() delivered the best results
## These have beem commented out to make the code block run in a sensible timeframe 

imputer = SimpleImputer() #set imputer
pca = PCA(15)
clf = svm.SVC() # define model 
pipe = Pipeline([('imp', imputer), ('scale', StandardScaler()), ('clf', clf)]) #create pipe 
# Parameters for gridsearch
param_grid ={
    'imp' : [imputer, IterativeImputer()],
    'scale' : [MinMaxScaler(), StandardScaler(), 'passthrough'],
    'clf__C': [0.1,1,10, 15, 20, 25,50, 100],
#     'pca__n_components': [2, 3, 5, 8, 10, 15],
#     'clf__kernel': ['rbf', 'linear', 'Sigmoid', 'Polynomial'],
    'clf__gamma' : [0.001, 0.01, 0.1, 0.15, 0.25, 0.5, 1]
} #set parameters for grid search 

grid = GridSearchCV(pipe, param_grid, cv=inner_cv, n_jobs=-1) # create grid search including pipe and parameters  
scores = cross_val_score(grid, x_train, y_train, scoring='accuracy', cv=outer_cv, n_jobs=-1) #nest grid inside CV
# report performance
print('Average cross_val score of best model: %.4f' % scores.mean())
```

## Logistic Regression 

```python
# set classifier
logistic = LogisticRegression(max_iter=10000, tol=0.1)
# Create pipe
pipe = Pipeline(steps=[('imp', imputer), ('scale', StandardScaler()), ('pca', pca), ('logistic', logistic)])
# Parameters for gridsearch
param_grid = {
    'imp' : [imputer, IterativeImputer()],
    'scale' : [MinMaxScaler(), StandardScaler(), 'passthrough'],
    'pca__n_components': [2, 3, 5, 8, 10, 15],
    'logistic__C': np.logspace(-4, 4, 4),
}

grid = GridSearchCV(pipe, param_grid, cv=inner_cv, n_jobs=-1) # create grid search including pipe and parameters 
# grid.fit(x_train, y_train) # fit 
scores = cross_val_score(grid, x_train, y_train, scoring='accuracy', cv=outer_cv, n_jobs=-1)
# # report performance
print('Average cross_val score of best model: %.4f' % scores.mean())
```

## Decision Tree 

```python
clf = DecisionTreeClassifier()
pipe = Pipeline(steps=[('imp', imputer),('DT', clf)])
# Parameters for gridsearch
param_grid = {
    'imp' : [imputer, IterativeImputer()],
    'DT__criterion': ['gini', 'entropy'],
    'DT__max_depth' : [2,4,6,8,10]
}
grid = GridSearchCV(pipe, param_grid, cv=inner_cv, n_jobs=-1) # create grid search including pipe and parameters 
scores = cross_val_score(grid, x_train, y_train, scoring='accuracy', cv=outer_cv, n_jobs=-1)
# # report performance
print('Average cross_val score of best model: %.4f' % scores.mean())
```
